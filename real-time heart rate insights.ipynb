```python
import matplotlib.pyplot as plt
import time
import numpy as np
i=1
j=[]
p=[]
while(1):
    result = firebase.get('https://medilink-96dd6-default-rtdb.asia-southeast1.firebasedatabase.app/py/users', None)
    time.sleep(1)
    p.append(result[0]["Heart_Rate"])
    j.append(i)
    i+=1
    if(i%200==0):
        plt.plot(j,p)
        plt.show()
        mean_hr = np.mean(p)
        max_hr = np.max(p)
        min_hr = np.min(p)
        std_hr = np.std(p)
        median_hr = np.median(p)
        print("Analysis and Insights:")
        print("Mean Heart Rate:", mean_hr, "bpm")
        print("Maximum Heart Rate:", max_hr, "bpm")
        print("Minimum Heart Rate:", min_hr, "bpm")
        print("Standard Deviation of Heart Rate:", std_hr, "bpm")
        print("Median Heart Rate:", median_hr, "bpm")
        break
```
